```python
# Ony Novianti (3A)
# Tugas Jobsheet 1

# IMPORT LIBRARY
import re
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# READ TEXT
text = open('orange.txt','r')
orange = text.read()
cleanString = orange.split(".")
cleanString = [s for s in cleanString if s != ':']
cleanString = [re.sub('\W+', " ", elem) for elem in cleanString]
cleanString = [i.strip() for i in cleanString]
# cleanString = [re.sub(r'[^A-Za-z]', '', elem) for elem in cleanString]

# cleanString = re.sub('\W+',' ', cleanString)
# cleanString=[cleanString] # supaya jadi string
# print(cleanString)
cleanString
```

```python
#TF-ID VECTORIZER
vectorizer = TfidfVectorizer(stop_words='english')
response = vectorizer.fit_transform(cleanString)
print(response)
```

```python
df = pd.DataFrame(
    response.todense(),
    columns = vectorizer.get_feature_names_out()
)
# pd.set_option('display.max_columns', None)
df.transpose()
```
